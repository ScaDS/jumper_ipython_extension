```python
import numpy as np

# Heavy CPU workload: dense matrix multiplication
size = 7000
A = np.random.rand(size, size)
B = np.random.rand(size, size)
C = A @ B
C = A @ C
print("CPU-bound test completed. Shape:", C.shape)
```
